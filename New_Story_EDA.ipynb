# Part 2:  Exploring Safety

After exploring the data, one of the things that jumped out to me was the amount of families that didn't feel safe in their previous homes.  Besides the typical necessities like food or water, being safe in one's own home is an essential human right that many of these families in Haiti do not have the fortune of having.  I decided to explore a little more into this issue which can be summed up in the graph below:

```python
# Families that have been robbed in the last year

families_robbed = len(baseline_df) - baseline_df['How many times were you stolen from?\nRobbery_Frequency'].isna().sum()

print('Stat: Families surveyed that have been robbed in the last year: {}'.format(families_robbed))
```

```python
countries = ('Colombia', 'U.S.', 'France', 'Haitians surveyed')
y_pos = np.arange(len(countries))
robberies = [90,490,592,6540]
 
plt.bar(y_pos, robberies, align='center', alpha=0.5)
plt.xticks(y_pos, countries)
plt.ylabel('Robberies', fontsize = 10)
plt.title('Robberies per 100,000 people', fontsize = 15)
 
plt.show();

# data from other countries come from : https://knoema.com/atlas/ranks/Burglary-rate
```

Now, obviously the graph is a little skewed because of the small number of people surveyed.  But of the 581 families surveyed, 38 of them reported AT LEAST one robbery in their home in the past year.  That is one order of magnitude greater than any country reported.  

### ...now onto the good news, (or, at least, the better news)

Now, since the graph above is based on the families BEFORE New Story became involved, I wanted to see how families' opinions on their safety changed after they moved into their new homes.  Below are the results:

```python
base_perc = [52, 53.3]
impact_perc = [95, 57.5]

ind = np.arange(len(base_perc))  
width = 0.25  

fig, ax = plt.subplots(figsize = (15, 10))
rects1 = ax.bar(ind - width/2, base_perc, width, color='IndianRed', label='Before')
rects2 = ax.bar(ind + width/2, impact_perc, width, color='RoyalBlue', label='After')


ax.set_ylabel('%', fontsize = 25)
ax.set_title('Safety Pre-Post Impact', fontsize = 25)
ax.set_xticks(ind)
ax.set_xticklabels(('Home', 'Community at night'), fontsize = 15)
# ax.set_yticks(fontsize = 15)
ax.legend(fontsize = 15);

```

After 6 months in their new homes, an overwhelming 95% of them felt safe in their new home, (versus only 52% in the baseline survey).  This is a huge improvement, and it's something that everyone at New Story can be proud of.  The community stats can still be improved on, however, and this is something that I would want to research further for future developments.
